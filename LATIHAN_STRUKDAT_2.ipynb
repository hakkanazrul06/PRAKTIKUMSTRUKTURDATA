```python
#satu
class JadwalNode:
  def __init__(self, nama_jadwal):
    self.nama_jadwal = nama_jadwal
    self.next = None

class JadwalList:
  def __init__(self):
    self.head = None

  def add_jadwal(self, nama_jadwal):
    jadwal_baru = JadwalNode(nama_jadwal)
    if not self.head:
      self.head = jadwal_baru
    else:
      current = self.head
      while current.next:
        current = current.next
      current.next = jadwal_baru

  def show_jadwal(self):
    if not self.head:
      print("\n Tidak ada jadwal hari ini!")
      return

    print("\n Daftar Jadwal Hari Ini:")
    current = self.head
    count = 1

    while current:
      print(f"{count}. {current.nama_jadwal}")
      current = current.next
      count += 1

list_jadwal = JadwalList()

npm = int(input("Masukkan NPM anda: "))

jumlah_jadwal = npm % 10

if jumlah_jadwal == 0:
  print("\nTidak ada jadwal hari ini!")
else:
  list_jadwal = JadwalList()

for i in range(1, (npm % 10) + 1):
  nama_jadwal = input(f"Masukkan jadwal ke-{i}: ")
  list_jadwal.add_jadwal(nama_jadwal)

list_jadwal.show_jadwal()
```

```python
#dua
class PageNode:
  def __init__(self, folder):
    self.folder = folder
    self.next = None
    self.prev = None

class FolderHistory:
  def __init__(self):
    self.current = None

  def visit_folder(self, folder):
      new_folder = PageNode(folder)
      if not self.current:
        self.current = new_folder

      else:
        new_folder.prev = self.current
        self.current.next = new_folder
        self.current = new_folder

  def show_history(self):
      if not self.current:
        print("\n File history is empty!")
        return

      print("\n File History:")
      temp = self.current
      while temp.prev:
        temp = temp.prev

      count = 1
      while temp:
        print(f"{count}. {temp.folder}")
        temp = temp.next
        count += 1

  def go_back(self):
      if self.current and self.current.prev:
        self.current = self.current.prev
        print(f"\n Anda kembali ke: {self.current.folder}")

      else:
        print("\n Tidak ada halaman sebelumnya!")

  def go_forward(self):
      if self.current and self.current.next:
        self.current = self.current.next
        print(f"\n Anda maju ke: {self.current.folder}")

      else:
        print("\n Tidak ada halaman berikutnya!")

history = FolderHistory()

npm = int(input("NPM: "))
jumlah = npm % 10

if jumlah == 0:
  print("Tidak ada.")
else:
  history = FolderHistory()

for i in range (1, (npm % 10) + 1):
  folder = input(f"Masukkan folder halaman ke-{i}: ")
  history.visit_folder(folder)

history.show_history()

while True:
  command = input("\n Tekan 'b' untuk kembali, 'f' untuk maju, tekan 'x' untuk keluar")

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari riwayat folder.")
    break
  else:
    print("\n Perintah tidak dikenal.")
```
